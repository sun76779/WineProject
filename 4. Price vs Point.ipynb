```python
%matplotlib inline
import pandas as pd
import matplotlib
import seaborn as sns
import csv
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

```python
wine_data_to_load = "/Users/Danny/Downloads/winemag-data-130k-v2 2.csv"

wine_data = pd.read_csv(wine_data_to_load)
```

```python
plt.figure(figsize=(12,6))
grouped = wine_data.groupby(["winery"]).mean().reset_index(drop=False).dropna() \
                    .sort_values('price',ascending=False).head(10)


sns.barplot(x='winery',y='price', data=grouped)
plt.show()
plt.savefig("SSENDTOSTEVEN.png")
```

```python
plt.figure(figsize=(12,6))
grouped = wine_data.groupby(["winery"]).mean().reset_index(drop=False).dropna() \
                    .sort_values('points',ascending=False).head(10)


sns.barplot(x='winery',y='points', data=grouped)
plt.show()
plt.savefig("/Users/Danny/Downloads/SSENDTOSTEVEN2.png")
```
